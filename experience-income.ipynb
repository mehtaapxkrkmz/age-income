```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import kagglehub
path = kagglehub.dataset_download("hussainnasirkhan/multiple-linear-regression-dataset")

print("Path to dataset files:", path)
```

```python
df=pd.read_csv("C:/Users/Asus/.cache/kagglehub/datasets/hussainnasirkhan/multiple-linear-regression-dataset/versions/1/multiple_linear_regression_dataset.csv")
df
```

```python
df = df.drop(columns=['age'])  
df
```

```python
import matplotlib.pyplot as plt

plt.scatter(df["experience"], df["income"], color="purple")
plt.xlabel("Tecrübe(Yıl)")
plt.ylabel("Maaş(TL)")
plt.title("Tecrübe-Maaş İlişkisi")
plt.show()
```

```python
# lineer regresyon modelini oluşturalım
from sklearn.linear_model import LinearRegression

model = LinearRegression()
```

```python
model.fit(df[['experience']],df['income'])
```

```python
tahminler = model.predict(df[['experience']])
tahminler
```

```python
model.coef_, model.intercept_
#modelin parametreleri.
```

```python
model.predict([[9]])
```

#örnek bir tahmin yaptık.9 yıllık tecrübe için maaş?
46374.81435644 TL 9 yıllık tecrübe için modelin tahmin ettiği maaş değeri.

```python
plt.scatter(df["experience"], df["income"], color="purple")
plt.plot(df["experience"], tahminler, color="turquoise")
plt.xlabel("Tecrübe(Yıl)")
plt.ylabel("Maaş(TL)")
plt.title("Tecrübe-Maaş İlişkisi")
plt.show()
```

Aşağıdaki fonksiyon, modelin tahmin ettiği değerlerle gerçek değerler arasındaki ilişkiyi değerlendirir ve genellikle R^2 skoru gibi bir ölçüm döner. Daha yüksek bir R^2 skoru, modelin verileri ne kadar iyi açıkladığını gösterir. Dolayısıyla, bu kod parçası aracılığıyla, modelin belirli bir özellik setine dayanarak gelir tahmini yapma başarısını ve uygunluğunu ölçmüş oluyoruz.

```python
model.score(df[["experience"]],df["income"])
```
